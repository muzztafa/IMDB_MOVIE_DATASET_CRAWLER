```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
```

```python
movie_name=[]
movie_url=[]
rating=[]
movie_rank=[]
movie_genre=[]
movie_genre_id=[]
movie_genre_year=[]
movie_genre_reviews=[]
i=0
movieId=2019
content_nx=[]
year=[]
movie_time=[]
views_count=[]
r_date_arr=[]
desc_fn=[]
no_reviews=[]
review_rate_arr=[]
review_title_arr=[]
reviewer_name_arr=[]
review_date_arr=[]
review_likes=[]
review_m_name=[]
review_count=[]
review_main_content=[]
atoi=0
driver = webdriver.Chrome('D:/IBA/data warehousing/mid_assignement/chromedriver_win32/chromedriver')
file = open('D:/IBA/data warehousing/mid_assignement/urls.txt', 'r')
Lines = file.readlines() 
for line in Lines:
    driver.get(line)
    content = driver.page_source
    soup = BeautifulSoup(content)
    time.sleep(10)
    rank=1
    for summary in soup.findAll('h3', attrs={'class':'lister-item-header'}):
        if(rank==2):
            break
        movie_url.append('https://www.imdb.com'+summary.a['href'])
        if soup.find('span', attrs={'class':'runtime'}) is not None:
            time_mv=soup.find('span', attrs={'class':'runtime'})
            s = time_mv.text.strip()
            result = re.search('(.*) min', s)
            if result is not None:
                movie_time.append(result.group(1))
            else:
                movie_time.append('N/A')
        else:
            movie_time.append('N/A')
        if soup.find('span', attrs={'name':'nv'}) is not None:
            views=soup.find('span', attrs={'name':'nv'})
            atoiViews=int(views.text.replace(',',''))
            views_count.append(atoiViews)
        else:
            views_count.append('N/A')
        driver.get('https://www.imdb.com'+summary.a['href'])
        content_nx .append(driver.page_source)
        data = BeautifulSoup(content_nx[i])
        time.sleep(10)
        if data.find('div', attrs={'class':'title_wrapper'}) is not None:
            movie=data.find('div', attrs={'class':'title_wrapper'})
            movie_name.append(movie.h1.text.strip())
        else:
            movie_name.append('N/A')
        movie_rank.append(rank)
        year.append(movieId)
        rank=rank+1
        if data.find('div', attrs={'class':'ratingValue'}) is not None:
            r=data.find('div', attrs={'class':'ratingValue'})
            rating.append(r.span.text)
        else:
            rating.append('N/A')
        if data.find('a', attrs={'title':'See more release dates'}) is not None:
            r_date=data.find('a', attrs={'title':'See more release dates'})
            r_date_arr.append(r_date.text)
        else:
            r_date_arr.append('N/A')
        if data.find('div', attrs={'class':'summary_text'}) is not None:
            desc=data.find('div', attrs={'class':'summary_text'})
            desc_fn.append(desc.text.strip())
        else:
            desc_fn.append('N/A')
        if data.find('div', attrs={'class':'user-comments'}) is not None:
            reviews=data.find('div', attrs={'class':'user-comments'})
            r_data=reviews.findAll('a')
            s = r_data[4].text.strip()
            result = re.search('See all (.*) user', s)
            if result is not None:
                a=result.group(1)
                atoi=int(a.replace(',',''))
                no_reviews.append(atoi)
            else:
                no_reviews.append('N/A')
        else:
            no_reviews.append('N/A')
        chk=0
        for item in data.findAll('div', attrs={'class':'see-more inline canwrap'}):
                if(item.h4==None):
                    break
                if(item.h4.text=='Genres:' or item.h4.text=='Genre:'):
                    for val in item.findAll('a'):
                        movie_genre_id.append(movie.h1.text.strip())
                        movie_genre.append(val.text)
                        movie_genre_year.append(movieId)
                        movie_genre_reviews.append(atoi)
                        chk=chk+1
                    break;
        if(chk==0):
            movie_genre_id.append(movie.h1.text)
            movie_genre.append('N/A')
        driver.get('https://www.imdb.com'+r_data[4]['href'])
        content_review =driver.page_source
        data_review = BeautifulSoup(content_review)
        time.sleep(10)
        for review_content in data_review.findAll('div', attrs={'class':'lister-item-content'}):
            review_m_name.append(movie.h1.text.strip())
            review_count.append(atoi)
            if review_content.find('span', attrs={'class':'rating-other-user-rating'}) is not None:
                review_rate_data=review_content.find('span', attrs={'class':'rating-other-user-rating'})
                review_rate_arr.append(review_rate_data.span.text)
            else:
                review_rate_arr.append('N/A')
            if review_content.find('a', attrs={'class':'title'}) is not None:
                review_title=review_content.find('a', attrs={'class':'title'})
                r_title=review_title.text.strip()
                review_title_arr.append(r_title)
            else:
                review_title_arr.append('N/A')
            if review_content.find('span', attrs={'class':'display-name-link'}) is not None:
                reviewer_name=review_content.find('span', attrs={'class':'display-name-link'})
                reviewer_name_arr.append(reviewer_name.a.text)
            else:
                reviewer_name_arr.append('N/A')
            if review_content.find('span', attrs={'class':'review-date'}) is not None:
                review_date=review_content.find('span', attrs={'class':'review-date'})
                review_date_arr.append(review_date.text)
            else:
                review_date_arr.append('N/A')
            if review_content.find('div', attrs={'class':'actions text-muted'}) is not None:
                review_rate=review_content.find('div', attrs={'class':'actions text-muted'})
                sa = review_rate.text.strip()
                result_lk = re.search('(.*) out', sa)
                if result_lk is not None:
                    review_likes.append(result_lk.group(1))
                else:
                    review_likes.append('N/A')
            else:
                review_likes.append('N/A')
            if review_content.find('div', attrs={'class':'text show-more__control'}) is not None:
                review_main=review_content.find('div', attrs={'class':'text show-more__control'})
                review_main_content.append(review_main.text)
            else:
                review_main_content.append('N/A')

        i=i+1
    
    df = pd.DataFrame({'Movie Name':review_m_name,'Reviewer':reviewer_name_arr,' Likes':review_likes,' Date of the review':review_date_arr,'Reviews Count':review_count,'Review Rating':review_rate_arr,'Review Title':review_title_arr,'Review':review_main_content}) 
    df.to_csv('D:/IBA/data warehousing/mid_assignement/movie_reviews'+str(movieId)+'.csv', index=False, encoding='utf-8')
    df = pd.DataFrame({'Movie Name':movie_genre_id,'Movie Genre':movie_genre,'Year':movie_genre_year,'Reviews Count':movie_genre_reviews}) 
    df.to_csv('D:/IBA/data warehousing/mid_assignement/movie_genre'+str(movieId)+'.csv', index=False, encoding='utf-8') 
    movieId=movieId-1

    
df = pd.DataFrame({'Movie Name':movie_name,'Rank':movie_rank,'Year':year,'Rating':rating,'Release Date':r_date_arr,' Commercial description':desc_fn,'Number of Reviews':no_reviews}) 
df.to_csv('D:/IBA/data warehousing/mid_assignement/movie.csv', index=False, encoding='utf-8')
 
```
